# Linear Regression


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/Concepts of Technology of AI/week7/advertising.csv')
data.info()
```

```python
x = data.drop('Sales',axis=1)
y = data['Sales']
```

```python
X_train, X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(Y_test,Y_pred)
r2 = r2_score(Y_test,Y_pred)
print(f"mean squared error: {mse}")
print(f"R-square: {r2}")
```

# Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/Concepts of Technology of AI/week7/heart.csv')
data.info()
data.head()
```

```python
x = data.drop('target',axis=1)
y = data['target'].values
```

```python
X_train, X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=2000)
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, Y_pred)
conf_matrix = confusion_matrix(Y_test, Y_pred)
class_report = classification_report(Y_test, Y_pred)
```

```python
print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix:\n{conf_matrix}")
print(f"Classification Report:\n{class_report}")
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('/content/drive/MyDrive/Concepts of Technology of AI/week7/diabetes.csv')
```

```python
data.info()
```

```python
x= data.drop(columns=['Outcome'])
y = data['Outcome']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
y_train
```

```python
tree = DecisionTreeClassifier(max_depth=5)
tree.fit(x_train, y_train)
```

```python
y_pred = tree.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
print(f"Custom Decision Tree Accuracy: {accuracy:.4f}")
```
